```python
import pandas as pd
import numpy as np
```

# Grouping data

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

## Which continent drinks more beer on average?

```python
drinks_more_beer_average = drinks.groupby(['continent']).mean(numeric_only=True)['beer_servings']
drinks_more_beer_average.idxmax()
```

## For each continent print the statistics for wine consumption

```python
drinks_continent_wine = drinks.groupby(['continent'])['wine_servings']
drinks_continent_wine.describe()
```

## Print the mean alcohol consumption per continent for every column

```python
drinks_alcohol_consumption_mean = drinks.groupby(['continent']).mean(numeric_only=True)
drinks_alcohol_consumption_mean
```

## Print the median alcohol consumption per continent for every column

```python
drinks_alcohol_consumption_median = drinks.groupby(['continent']).median(numeric_only=True)
drinks_alcohol_consumption_median
```

## Print the mean, min and max values for spirit consumption

```python
drinks_continent_wine = drinks.groupby(['continent'])['spirit_servings']
drinks_continent_wine.describe().loc[:,['mean','min','max']]

# alternative
drinks.groupby('continent').spirit_servings.agg(['mean', 'min', 'max'])
```

# Pivoting

```python
df = pd.read_csv('../data/titanic.csv')
df.head()
```

## Find average survival rate by gender

```python
df.groupby(['sex']).mean(numeric_only=True).survived
```

## Create a Pivot table and find survival rate by gender on various classes

```python
df.pivot_survivedtable('survived', index='sex', columns='class')
```

## Create a Pivot table and find survival rate by gender, age wise of various classes

```python
df.pivot_table('survived', index=['sex','age'], columns='class')
```

## Partition each of the passengers into four categories based on their age (age categories (0, 10), (10, 30), (30, 60), (60, 80)) and then create a pivot table that counts survival by gender, categories wise age of various classes

```python
# pd.cut()
# pivot_table()
ages = [0,10,30,60,80]
categories = pd.cut(df.loc[:,'age'],ages)

df.pivot_table('survived', index=['sex',categories], columns='class', aggfunc=np.sum)
```
