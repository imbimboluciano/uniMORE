```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn import preprocessing
```

```python
# 1. Download the dataset from https://archive.ics.uci.edu/ml/datasets/mushroom and insert it into the dataframe `mushroom`.
mushroom = pd.read_csv("data/agaricus-lepiota.data", sep=',', header=None)
mushroom_columns = ["target","cap-shape","cap-surface","cap-color","bruises","odor","gill-attachment","gill-spacing","gill-size","gill-color","stalk-shape","stalk-root",
                    "stalk-surface-above-ring","stalk-surface-below-ring","stalk-color-above-ring","stalk-color-below-ring","veil-type","veil-color","ring-number",
                    "ring-type","spore-print-color","population","habitat"]
mushroom.columns = mushroom_columns
mushroom
```

```python
# 2. Transform the attribute values into numbers.
enc = preprocessing.OrdinalEncoder() # ordinal because we don't have boolean value
enc.fit(mushroom)
mushroom_encoded = pd.DataFrame(enc.transform(mushroom), columns=mushroom_columns)
mushroom_encoded
```

```python
# 3. Calculate the accuracy (5-fold cross-validation) of the obtained dataset in predicting whether a mushroom is poisonous using 3 classifiers.
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, train_test_split
```

```python
y = mushroom_encoded['target']
X = mushroom_encoded.loc[:,'cap-shape':'habitat']
```

```python
lr = LogisticRegression()
scores = cross_val_score(lr,X,y, cv = 5)
print(scores)
scores.mean()
```

```python
svm = SVC(C=0.6)
scores = cross_val_score(svm,X,y, cv = 5)
print(scores)
scores.mean()
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
scores = cross_val_score(knn,X,y, cv = 5)
print(scores)
scores.mean()
```

```python
# 4. Use only the attributes of the dataset that pertain to color (call the new dataset `mushroom_color`) and check if the accuracy changes.
mushroom_color = pd.DataFrame()

for column_name in mushroom.columns:
    if column_name.__contains__("color"):
        mushroom_color[column_name] = mushroom_encoded[column_name]

mushroom_color
```

```python
lr = LogisticRegression()
scores = cross_val_score(lr,mushroom_color,y, cv = 5)
print(scores)
scores.mean()
```

```python
svm = SVC(C=0.6)
scores = cross_val_score(svm,mushroom_color,y, cv = 5)
print(scores)
scores.mean()
```

```python
knn = KNeighborsClassifier(n_neighbors=4)
scores = cross_val_score(knn,mushroom_color,y, cv = 5)
print(scores)
scores.mean()
```
