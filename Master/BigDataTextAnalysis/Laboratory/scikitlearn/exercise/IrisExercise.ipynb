```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
from sklearn import datasets
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import FunctionTransformer
```

```python
# 1. Create a dataframe for iris dataset

iris = datasets.load_iris()
iris['feature_names']
iris['target_names']

X = iris['data']
y = iris['target']
y

df = pd.DataFrame(X, columns=iris['feature_names'])
df.head(10)
```

```python
# 2.Add two new attributes: area of petal and and area of sepal 
# Compute accuracy with train /test 0.75/0.25 and a Decision Tree Classifier

df["petal area (cm2)"] = df["petal length (cm)"] * df["petal width (cm)"]
df["sepal area (cm2)"] = df['sepal length (cm)'] * df['sepal width (cm)']
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df,y, test_size=0.25)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

result = dt.predict(X_test)
accuracy_score(result, y_test)
```

```python
accuracy_score(y_train, dt.predict(X_train)) # accuracy train
```

```python
# 3. Create a discretization of the values into 4 bins (for each column of the dataset),
# replace the initial dataset with the discretized one, and calculate the accuracy

vbins = preprocessing.KBinsDiscretizer(n_bins=4, encode='ordinal').fit(df)
new_df = pd.DataFrame(vbins.transform(df), columns=df.columns)
new_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(new_df,y, test_size=0.25)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

result = dt.predict(X_test)
accuracy_score(y_test, result) # accuracy test
```

```python
accuracy_score(y_train, dt.predict(X_train)) # accuracy train
```

```python
# 4. Standardize the dataset from step 2 and calculate the accuracy with the previous classifier.
df_scaled = pd.DataFrame(preprocessing.scale(df), columns=df.columns)
df_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_scaled,y, test_size=0.25)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

result = dt.predict(X_test)
accuracy_score(result, y_test)
```

```python
accuracy_score(y_train, dt.predict(X_train)) # accuracy train
```

```python
# 5. Apply the logarithm (custom transformation) to all columns of the dataset from step 2 and calculate the accuracy with the previous classifier.
transformer = FunctionTransformer(np.log1p, validate=True)
transformed_df = pd.DataFrame(transformer.transform(df), columns=df.columns)
transformed_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(transformed_df,y, test_size=0.25)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

result = dt.predict(X_test)
accuracy_score(result, y_test)
```

```python
accuracy_score(y_train, dt.predict(X_train)) # accuracy train
```

```python
# 6. Create a new column that has a value of 1 if and only if both the petal area and the sepal area are greater than 4, 
# and calculate the accuracy with the previous classifier.

def checkarea(X):
    mask0 = X[:,0] > 4 # petal area
    mask1 = X[:,1] > 4 # sepal area
    return mask0 & mask1

checkarea(X)

transformer = FunctionTransformer(checkarea, validate=True)
df["bool_area"] = transformer.transform(df.loc[:,['petal area (cm2)','sepal area (cm2)']].values)
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(transformed_df,y, test_size=0.25)
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

result = dt.predict(X_test)
accuracy_score(result, y_test)
```

```python
accuracy_score(y_train, dt.predict(X_train)) # accuracy train
```
